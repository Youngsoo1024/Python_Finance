```python
import random
I= 5000
```

```python
%time mat = [[random.gauss(0,1) for j in range(I)]\
             for i in range(I)]
```

```python
mat[0][:5]
```

```python
%time sum ([sum(l) for l in mat])
# the sums of the single list objects are first calculated during 
# a list comprehension; then the sum of the sums is taken
```

```python
import sys
sum([sys.getsizeof(l) for l in mat])
# This adds up thememory usage of all list objects
```
