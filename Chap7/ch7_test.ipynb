```python
import numpy as np
```

```python
np.random.seed(1000)
```

```python
y = np.random.standard_normal(20)
```

```python
x = np.arange(len(y))
plt.plot(x,y)
```

```python
plt.plot(y)
```

```python
plt.plot(y.cumsum())
```

```python
plt.plot(y.cumsum())
plt.grid(False) # Turn off grid
plt.axis('equal') # Leads to equal scaling for the two axes.
```

```python
plt.plot(y.cumsum())
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1,
         np.max(y.cumsum())+1)
```

```python
plt.figure(figsize=(10, 6))  
plt.plot(y.cumsum(), 'b', lw=1.5)  
plt.plot(y.cumsum(), 'ro')  
plt.xlabel('index')  
plt.ylabel('value')  
plt.title('A Simple Plot'); 
```

```python
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y, lw=1.5)
plt.plot(y, 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot');
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw=1.5, label='1st')  
plt.plot(y[:, 1], lw=1.5, label='2nd')  
plt.plot(y, 'ro')
plt.legend(loc=0)  
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```
