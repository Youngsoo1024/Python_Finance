```python
import numpy as np
import pandas as pd
from pylab import mpl, plt
plt.style.use('seaborn-v0_8-paper')
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
filename = 'tr_eikon_eod_data.csv'  
```

```python
f = open(filename, 'r')
f.readlines()[:5]
```

```python
data = pd.read_csv(filename,  
                           index_col=0, 
                           parse_dates=True) 
```

```python
data.head()
```

```python
data.tail()
```

```python
data.plot(figsize=(10,12), subplots=True)
```

```python
instruments = ['Apple Stock', 'Microsoft Stock',
                       'Intel Stock', 'Amazon Stock', 'Goldman Sachs Stock',
                       'SPDR S&P 500 ETF Trust', 'S&P 500 Index',
                       'VIX Volatility Index', 'EUR/USD Exchange Rate',
                       'Gold Price', 'VanEck Vectors Gold Miners ETF',
                       'SPDR Gold Trust']
```

```python
for ric, name in zip(data.columns, instruments):
    print('{:8s}|{}'.format(ric, name))
```

```python
data.aggregate([min,  
                         np.mean,  
                         np.std,  
                         np.median,  
                         max]  
         ).round(2)
```

```python
data.diff().head() 
```

```python
data.diff().mean()
```

```python
data.pct_change().round(3).head()
```

```python
data.pct_change().mean().plot(kind='bar', figsize=(10,6))
```

Figure 8-2. Mean values of percentage changes as bar plot


```python
rets = np.log(data/data.shift(1))
```

```python
rets.head().round(3)
```

```python
rets.cumsum().apply(np.exp).plot(figsize=(10,6))
```

Figure 8-3. Cumulative log returns over time


```python
data.resample('W', label='right').last().head()

#1w will be depreciated use W instead
# weakly time interval
```

```python
data.resample('ME', label='right').last().head()
#monthly time intervals
```

```python
rets.cumsum().apply(np.exp).resample('ME',label='right').last().plot(figsize=(10,6))
```

8-4. Resampled cumulative log returns over time(monthly)
